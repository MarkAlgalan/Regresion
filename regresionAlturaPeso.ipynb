```python
import pandas as pd
datos = pd.read_csv("alturas-pesos.csv")
```

El archivo fue obtenido de https://github.com/lab-ml-itba/Regresion-Lineal/blob/master/alturas-pesos.csv

```python
datos.info()
datos.tail()
```

```python
datos.head()
```

```python
import seaborn as sb

sb.scatterplot(x="Altura",y="Peso",data=datos,hue="Peso",palette="Blues")
```

```python
x = datos["Altura"]
y = datos["Peso"]
```

```python
x_proc = x.values.reshape(-1,1)
y_proc = y.values.reshape(-1,1)
```

```python
from sklearn.linear_model import LinearRegression
modelo = LinearRegression()
```

```python
modelo.fit(x_proc,y_proc)
```

```python
altura = 172
modelo.predict([[altura]])
```

```python
modelo.score(x_proc,y_proc)
```

Pues el modelo es una aproximación a la realidad y al ser el peso de una persona algo muy variable puede que muchas veces no sea el correcto pero no se alejara tanto más alla de unos datos anomalas. Recordar que estos modelos no representan en su totalidad la realidad pero si muestran una clara relación en este caso entre edad y altura al haber varios datos es más preciso. En el lado negativo es que no prueba con nuevos datos.
